```python
def cont_disc(df):
    continuous = pd.DataFrame()
    discrete = pd.DataFrame()
    for col in df:
        if df[col].nunique() > 15:
            continuous[col] = df[col]
        elif df[col].nunique() <= 15:
            discrete[col] = df[col]
    return [continuous, discrete]
```

```python
continuous=cont_disc(customer_df)[0]
discrete=cont_disc(customer_df)[1]
```

```python
def substitution(df1,df2):
    for col in df2:
        df1[col]=df2[col]
    return df1
```

```python
X_cont=cont_disc(X_train_num)[0]
X_disc=cont_disc(X_train_num)[1]
X_cont.describe().T
```

```python
def remove_outliers(df):
    df2=df.copy()
    for col in df.columns:
        iqr = np.percentile(df2[col],75) - np.percentile(df2[col],25)
        upper_limit = np.percentile(df2[col],75) + 3*iqr
        lower_limit = np.percentile(df2[col],25) - 3*iqr
        df2 = df2[(df2[col]>lower_limit) & (df2[col]<upper_limit)]
    return df2
```

```python
no_out=remove_outliers(X_cont)
no_out.describe().T
```

```python
X_train_num=pd.concat([no_out, X_disc], axis=1)
X_train_num=X_train_num.dropna(axis=0)
X_train_num
```

```python
X_train=pd.concat([X_train_num, X_train_cat], axis=1)
X_train=X_train.dropna(axis=0)
X_train
```

```python
X_train2=X_train.copy()
X_train_num = X_train2.select_dtypes(include = np.number)
X_train_cat = X_train2.select_dtypes(include = object)
X_test_num = X_test.select_dtypes(include = np.number)
X_test_cat = X_test.select_dtypes(include = object)
print(X_train_num.shape)
print(X_train_cat.shape)
print(X_test_num.shape)
print(X_test_cat.shape)
```

```python
X_train_cat_ord = X_train_cat[['coverage','education','employmentstatus','location_code','vehicle_size']]
X_train_cat_nom = X_train_cat[['state','response','gender','marital_status','policy_type','renew_offer_type','sales_channel','vehicle_class']]
X_test_cat_ord = X_test_cat[['coverage','education','employmentstatus','location_code','vehicle_size']]
X_test_cat_nom = X_test_cat[['state','response','gender','marital_status','policy_type','renew_offer_type','sales_channel','vehicle_class']]
display(X_train_cat_ord.shape)
display(X_train_cat_nom.shape)
display(X_test_cat_ord.shape)
display(X_test_cat_nom.shape)
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
```

```python
ohe=OneHotEncoder(drop='first')
ohe.fit(X_train_cat_nom)
train_nom_encoded = ohe.transform(X_train_cat_nom).toarray()
train_nom_cat_encoded = pd.DataFrame(train_nom_encoded,columns=ohe.get_feature_names_out())
train_nom_cat_encoded.head()
```

```python
test_nom_encoded = ohe.transform(X_test_cat_nom).toarray()
test_nom_cat_encoded = pd.DataFrame(test_nom_encoded,columns=ohe.get_feature_names_out())
test_nom_cat_encoded.head()
```

```python
orde=OrdinalEncoder()
orde.fit(X_train_cat_ord)
train_ord_encoded = orde.transform(X_train_cat_ord)
train_ord_cat_encoded = pd.DataFrame(train_ord_encoded,columns=orde.get_feature_names_out())
train_ord_cat_encoded.head()
```

```python
test_ord_encoded = orde.transform(X_test_cat_ord)
test_ord_cat_encoded = pd.DataFrame(test_ord_encoded,columns=orde.get_feature_names_out())
test_ord_cat_encoded.head()
```

```python
X_train_transformed=pd.concat([X_train_norm, train_nom_cat_encoded, train_ord_cat_encoded], axis=1)
X_train_transformed
```

```python
X_test_transformed=pd.concat([X_test_norm, test_nom_cat_encoded, test_ord_cat_encoded], axis=1)
X_test_transformed
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
```

```python
lr = LinearRegression().fit(X_train_transformed,y_train)
lr_score = lr.score(X_test_transformed, y_test)
print("The R2 of the linear regression model is {:.2f}".format(lr_score))
```

```python
def train_models (list_of_models, X_train, y_train):
    trained_models=[]
    for model in (list_of_models):
        print('training {} model'.format(model))
        model.fit(X_train, y_train)
        trained_models.append(model)
    return trained_models
```

```python
def evaluate_models(list_of_trained_models, X_test, y_test):
    score=[]
    for model in (list_of_trained_models):
        score.append(model.score(X_test, y_test))
    for i in range(len(list_of_trained_models)):
        print(f'The {list_of_trained_models[i]} model has a score of {score[i]}')
    return score
```

```python
models=[LinearRegression(), KNeighborsRegressor(n_neighbors=4), KNeighborsRegressor(), KNeighborsRegressor(n_neighbors=6)]
trained_models=train_models(models, X_train_transformed, y_train)
scores=evaluate_models(trained_models, X_test_transformed, y_test)
```

```python
models2=[KNeighborsRegressor(n_neighbors=6), KNeighborsRegressor(n_neighbors=7), KNeighborsRegressor(n_neighbors=8)]
trained_models2=train_models(models2, X_train_transformed, y_train)
scores2=evaluate_models(trained_models2, X_test_transformed, y_test)
```

```python
from sklearn.neural_network import MLPRegressor
new_models=[KNeighborsRegressor(n_neighbors=6, weights ='distance'), MLPRegressor()]
new_trained_models=train_models(new_models, X_train_transformed, y_train)
scores=evaluate_models(new_trained_models, X_test_transformed, y_test)
```

It is extremely convenient to have the functions that train and evaluate models, as shown when we optimized the neighbors for the knn and afterwards trained a completely new type of regressor model, the MLPRegressor, which behaved better than even a semi-optimized KNN. As expected the Linear model had the worst score, and we managed to find the amount of optimal neighbors for the KNN, which is 6, and that weighted contributions also increased the score.
